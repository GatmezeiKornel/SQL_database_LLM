```python
filename = "GYOGYSZ.csv"
df = pd.read_csv(f"forras/csv/{filename}", sep=';')
df
```

```python
clean_df = df.dropna(axis=1,how='all')
clean_df
```

```python
compare=df.columns.difference(clean_df.columns)
compare
```

```python
clean_df.columns
```

```python
clean_df = clean_df.drop(axis=1,columns=["SWITCH","OEP_AFA","NORM_MIHAID","NORM_MIHASTAT"])
clean_df
```

```python
idxs=[]
for idx, value in clean_df.iterrows():
    if str(value['TKTORLESDAT']) != 'nan':
        idxs.append(idx)
```

```python
len(idxs)
```

```python
row_clean_df = clean_df.drop(axis=0,index=idxs)
```

```python
row_clean_df
```

```python
row_clean_df = row_clean_df.drop(axis=1,columns=["TKTORLES","TKTORLESDAT"])
```

```python
row_clean_df
```

```python
row_clean_df.to_csv(f"forras/csv/{filename}", sep=';',index=False)
```
